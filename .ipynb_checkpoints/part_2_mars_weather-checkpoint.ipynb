# Mars Weather

```python
# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import datetime as dt
from datetime import timedelta
from datetime import datetime
```

```python
browser = Browser('chrome')
```

## Step 1: Visit the Website

```python
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"

browser.visit(url)
```

## Step 2: Scrape the Table

```python
# Scrape the website
html = browser.html

# Create a BeautifulSoup object from scraped HTML
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Show scraped HTML
print(soup.prettify())
```

```python
# Find the Mars Temperature Data table
table = soup.find('table', class_='table')
table
```

## Step 3: Store the Data

```python
# Find table headers, create empty list, loop through then append
headers = table.find_all('th')

table_headers=[]

for header in headers:
    column_header = header.text
    table_headers.append(column_header)
    
table_headers
```

```python
# Create Pandas DataFrame
mars_df = pd.DataFrame(columns=table_headers)
mars_df
```

```python
# Extract the table rows and place into DataFrame
for row in soup.find_all('tr',class_='data-row'):
        data = row.find_all('td')
        row_data = [td.text.strip() for td in data]
        length = len(mars_df)
        mars_df.loc[length] = row_data
        
mars_df.head()
```

```python
# Alternate Method: 
# Using Pandas to read HTML URL and place into table
table = pd.read_html(url)

# Convert table into Pandas DataFrame
alt_mars_df = table[0]
alt_mars_df.head()
```

## Step 4: Prepare Data for Analysis

```python
# Obtain Mars_df information to examine data types for each column
mars_df.info()
```

```python
# Convert each Column into the appropriate Data Type
mars_df['id'] = mars_df['id'].astype(int)
mars_df['terrestrial_date'] = pd.to_datetime(mars_df['terrestrial_date'])
mars_df['sol'] = mars_df['sol'].astype(int)
mars_df['ls'] = mars_df['ls'].astype(int)
mars_df['month'] = mars_df['month'].astype(int)
mars_df['min_temp'] = mars_df['min_temp'].astype(float)
mars_df['pressure'] = mars_df['pressure'].astype(float)
```

```python
# Look at Mars_df data types for each column again
mars_df.info()
```

## Step 5: Analyze the Dataset

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.

**1. How many months exist on Mars?**

```python
# Examine how many months exist on Mars?
months_on_mars = mars_df['month'].nunique()
months_on_mars
```

```python
# Print Statement
print(f"{months_on_mars} months exist on Mars.")
```

**2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?**

```python
# Find min(first Martian day) and max(last Martian day) and then total unique days in the dataset
first_martian_day = mars_df['sol'].min()
print(first_martian_day)

last_martian_day = mars_df['sol'].max()
print(last_martian_day)

unique_martian_days = mars_df['sol'].nunique()
print(unique_martian_days)
```

```python
# Print Statement
print(f"There are {unique_martian_days} Martian days worth of data in the dataset. The days range from\
 day {first_martian_day} to day {last_martian_day}.")
```

**3. What are the coldest and warmest months on Mars (at the location of Curiosity)?**

```python
# Find average minimum daily temperatures for all months
avg_temps_month_df = mars_df.groupby('month', as_index=True).agg({'min_temp': np.mean})
avg_temps_month_df
```

```python
# Find Coldest Month

# First find the coldest temperature
coldest_temp = avg_temps_month_df['min_temp'].min()
rnd_coldest_temp = round(coldest_temp, 2)
print(rnd_coldest_temp)

# Use coldest temperature to find the proper row and associated index
coldest_month = avg_temps_month_df.loc[avg_temps_month_df['min_temp'] == coldest_temp].index[0]
print(coldest_month)
```

```python
# Print Statement
print(f"The coldest Martian Month is {coldest_month} with a temperature of {rnd_coldest_temp} degrees.")
```

```python
# Find Warmest Month

# First find the warmest temperature
warmest_temp = avg_temps_month_df['min_temp'].max()
rnd_warmest_temp = round(warmest_temp, 2)
print(rnd_warmest_temp)

# Use warmest temperature to find the proper row and associated index
warmest_month = avg_temps_month_df.loc[avg_temps_month_df['min_temp'] == warmest_temp].index[0]
print(warmest_month)
```

```python
# Print Statement
print(f"The warmest Martian Month is {warmest_month} with a temperature of {rnd_warmest_temp} degrees.")
```

```python
# Plot results as a bar chart

avg_temps_month_df.plot(kind='bar')
plt.title('Mars Average Daily Temperatures per Martian Month')
plt.ylabel('Temperature')
plt.xlabel('Martian Month')
plt.tight_layout()
plt.show()
```

**4. Which months have the lowest and the highest atmospheric pressure on Mars?**

```python
# Find average daily atmospheric pressures for all months
avg_atmpressure_month_df = mars_df.groupby('month', as_index=True).agg({'pressure': np.mean})
avg_atmpressure_month_df
```

```python
# Find Lowest Atmospheric Pressure 

# First find the lowest pressure
lowest_atmpressure = avg_atmpressure_month_df['pressure'].min()
rnd_lowest_atmpressure = round(lowest_atmpressure,2)
print(rnd_lowest_atmpressure)

# Use lowest pressure to find the proper row and associated index
lowest_atmpressure_month = avg_atmpressure_month_df.loc\
    [avg_atmpressure_month_df['pressure'] == lowest_atmpressure].index[0]
print(lowest_atmpressure_month)
```

```python
# Print Statement
print(f"The month with the lowest atmospheric pressure is Martian Month {lowest_atmpressure_month}\
 with a pressure of {rnd_lowest_atmpressure}.")
```

```python
# Find Highest Atmospheric Pressure 

# First find the highest pressure
highest_atmpressure = avg_atmpressure_month_df['pressure'].max()
rnd_highest_atmpressure = round(highest_atmpressure,2)
print(rnd_highest_atmpressure)

# Use highest pressure to find the proper row and associated index
highest_atmpressure_month = avg_atmpressure_month_df.loc\
    [avg_atmpressure_month_df['pressure'] == highest_atmpressure].index[0]
print(highest_atmpressure_month)
```

```python
# Print Statement
print(f"The month with the highest atmospheric pressure is Martian Month {highest_atmpressure_month}\
 with a pressure of {rnd_highest_atmpressure}.")
```

```python
# Plot results as a bar chart

avg_atmpressure_month_df.plot(kind='bar')
plt.title('Mars Average Atmospheric Pressure per Martian Month')
plt.ylabel('Atmospheric Pressure')
plt.xlabel('Martian Month')
plt.tight_layout()
plt.show()
```

**5. How many terrestrial (Earth) days exist in a Martian Year?**

```python
# Visually estimate the result by plotting the daily minimum temperature
mars_df['min_temp'].plot()
plt.title('Estimated Terrestrial Days in a Martian Year')
plt.xlabel('Number of Terrestrial Days')
plt.ylabel('Minimum Temperature')
plt.tight_layout()
plt.show()


print(f"The distance from peak to peak is roughly 1425-750, or 675 days. A year on Mars appears to be about\
 675 days from the plot. Internet search confirms that a Mars year is equivalent to 687 earth days.")
```

```python
# Corroborated above with coding below:
# Start by finding the initial solar longitude. This is anticipated to be the same ~1 Martian year later.
initial_sol_lng = mars_df['ls'].loc[0]


# Find the rows where the solar longitude is equivalent to the initial solar longitude. Parse through rows. 
equal_sol_lngs = mars_df.loc[mars_df['ls'] == initial_sol_lng]
equal_sol_lngs
```

```python
# Find the date of the initial solar longitude
date_initial_sol_lng = equal_sol_lngs['terrestrial_date'].loc[0]
# print(date_initial_sol_lng)

# Find the date one year after the initial solar longitude
date_one_year_martian_year = equal_sol_lngs['terrestrial_date'].iloc[1]
# print(date_one_year_martian_year)

# To find one Martian year
one_martian_year = date_one_year_martian_year - date_initial_sol_lng

print(f"There are {one_martian_year} on Earth for one Martian year, which is similar to the Internet search of\
 687 Earth days is equivalent to one Martian year.")
```
